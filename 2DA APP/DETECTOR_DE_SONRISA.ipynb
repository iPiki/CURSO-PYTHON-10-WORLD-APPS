```python
import cv2
smile_cascade=cv2.CascadeClassifier("haarcascade_smile_2.xml")
imagen=cv2.imread("escalada_varias_sonrisas_1.jpg")
sonrisas=smile_cascade.detectMultiScale(imagen,
scaleFactor=1.1,
minNeighbors=5,
minSize=(10,10),
maxSize=(50,40))
for x , y, w, h in sonrisas:
    img=cv2.rectangle(imagen,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow("Imagen normal",imagen)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```
