```python
import cv2
video=cv2.VideoCapture("videoprueba.mp4")
a=0
while True:
    a=a+1
    check,frame=video.read()
    print(check)
    print(frame)
    gris=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Captura",gris)
    key=cv2.waitKey(10)
    if key==ord('q'):
        break
print(a)
video.release()
cv2.destroyAllWindows()
```
